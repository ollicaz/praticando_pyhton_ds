# Atividade 2

```python
# Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para visualizar o dicionário contendo as chaves de grupos formados e a lista de índices de cada grupo.

indices = emissoes_gases.groupby('Nível 1 - Setor').groups
indices
```

```python
# Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" e localize os dados do grupo "Agropecuária".

agropecuaria = emissoes_gases.groupby('Nível 1 - Setor').get_group('Agropecuária')
agropecuaria
```

```python
emissoes_gases[2021]
```

```python
# Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a média de emissão de cada setor no ano de 2021

media_emissao = emissoes_gases.groupby('Nível 1 - Setor').mean(2021)
media_emissao
```

```python
# Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a soma de emissão de cada setor. Ordene os dados da maior para menor emissão.
#primeiro passo juntar todos os anos:
colunas_info = list(emissoes_gases.loc[:,'Nível 1 - Setor':'Produto'].columns)
anos_info = list(emissoes_gases.loc[:,1970:2021].columns)
emissoes_por_ano = emissoes_gases.melt(id_vars=colunas_info,value_vars=anos_info,var_name='Ano',value_name='Emissão')

#agora que todos os valores de emissão estão em uma coluna, utilizando o comando .sum()
emissoes_por_ano.groupby(['Nível 1 - Setor','Atividade Econômica'])[['Emissão']].sum().sort_values('Emissão',ascending=False)

# emissao_setores = emissoes_gases.groupby('Nível 1 - Setor').sum(numeric_only=True).sum(axis=1).sort_values(ascending=False)
# emissao_setores
```

# Atividade 3

```python
# A equipe de supervisão deseja saber agora a média de emissão para cada um dos setores econômicos da coluna "Nível 1 - Setor"
# ao longo dos anos. Como desafio, crie uma tabela de pivô e uma visualização gráfica contendo essas informações, utilizando apenas
# o método pivot_table para a construção da tabela, sem utilizar o método groupby.


emissao_setores = emissoes_por_ano.pivot_table(values = 'Emissão', index = 'Ano', columns = 'Nível 1 - Setor', aggfunc = 'mean')
emissao_setores
```

# Atividade 4

```python
#Faça um agrupamento de dados com as colunas "Estado" e "Nível 1 - Setor", obtendo a soma de emissão e armazenando o resultado em uma tabela.

emissoes_estado_setor = emissoes_gases.groupby(['Estado', 'Nível 1 - Setor']).sum('Emissão')
emissoes_estado_setor
```

```python
# Utilizando a tabela construída na atividade 1, selecione os dados referentes à "Energia" do índice "Nível 1 - Setor".

emissoes_estado_setor.xs('Energia', level = 1)
```

```python
# Utilizando a tabela construída na atividade 1, encontre a atividade econômica com valor máximo de emissão do Estado de Minas Gerais

emissoes_estado_setor.xs('MG', level = 0).idxmax()
```

```python
# Obtenha uma tabela contendo a atividade econômica com máxima emissão para cada Estado.

emissoes_estado_setor.groupby(level= 0).idxmax()
```

```python
# Obtenha uma tabela contendo o Estado com máxima emissão para cada atividade econômica.

emissoes_estado_setor.groupby(level= 1).idxmax()
```
