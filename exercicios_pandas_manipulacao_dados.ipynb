# Projeto 1 - Vendas online

```python
import pandas as pd
import numpy as np
```

```python
dados_vendas = pd.read_json('https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json')
dados_vendas
```

```python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
dados_vendas
```

```python
# Remover os dados em listas dentro do DataFrame;
# Verificar os tipos de dados;
# Identificar colunas numéricas;
# Transformar a coluna numérica para o tipo numérico.
```

```python
colunas = list(dados_vendas)
colunas
```

```python
dados_vendas = dados_vendas.explode(colunas[1:])
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas.reset_index(inplace = True, drop = True)
dados_vendas
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '').replace('R', '').strip())
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].astype(np.float64)
dados_vendas
```

```python
dados_vendas.info()
```

```python
# Devido a uma instabilidade no site da empresa, tivemos problemas com os nomes dos clientes durante o salvamento.
# Isso resultou em uma coluna de nomes de clientes com uma mistura de letras, maiúsculas e minúsculas, números e outros caracteres.

# Sabendo disso, manipule os textos presentes na coluna Cliente para que seja obtido como resultado os nomes dos clientes em letras minúsculas,
# com a ausência de caracteres especiais ou números.
```

```python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.replace('[^a-zA-Z\-\']', '', regex = True)
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.replace('(?<!\w)-(?!\w)', '', regex = True)
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower()
dados_vendas['Cliente']
```

```python
dados_vendas.head()
```

```python
# O objetivo desse projeto é realizar uma análise dos resultados de um evento com os clientes de uma empresa de vendas online.
# Foi coletado um conjunto de dados que contém os clientes que mais gastaram com produtos dentro de 5 dias de vendas, que é o período de duração do evento.
# Essa análise vai identificar o cliente com a maior compra na semana, que irá receber um prêmio da loja, e posteriormente, ela pode ajudar
# a empresa a criar novas estratégias para atrair mais clientes.
```

```python
dados_vendas['Data de venda'] = pd.to_datetime(dados_vendas['Data de venda'])
```

```python
total_compra = dados_vendas.groupby(dados_vendas['Data de venda'].dt.strftime('%d-%m'))['Valor da compra'].sum()
total_compra
```

# Projeto 2 - Administração de condomínios

```python
dados_condominio = pd.read_json('https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json')
dados_condominio
```

```python
dados_condominio = pd.json_normalize(dados_condominio['dados_locacao'])
dados_condominio
```

```python
# Remover os dados em listas dentro do DataFrame;
# Verificar os tipos de dados;
# Identificar colunas numéricas;
# Transformar a coluna numérica para o tipo numérico.
```

```python
colunas2 = list(dados_condominio)
colunas2
```

```python
dados_condominio = dados_condominio.explode(colunas2[1:])
dados_condominio.reset_index(inplace = True, drop = True)
dados_condominio
```

```python
dados_condominio['valor_aluguel'] = dados_condominio['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',', '').replace('reais', '').strip())
dados_condominio['valor_aluguel'] = dados_condominio['valor_aluguel'].astype(np.float64)
dados_condominio
```

```python
dados_condominio.info()
```

```python
# manipule os textos na coluna apartamento para remover o texto (blocoAP) do DataFrame.
```

```python
dados_condominio['apartamento'] = dados_condominio['apartamento'].str.replace('[(blocoAP)]', '', regex = True)
```

```python
dados_condominio.head()
```

```python
# A administração de condomínios é uma tarefa que requer muita atenção e organização.
# Entre as diversas responsabilidades da gestão está o recebimento dos aluguéis dos locatários. Para garantir uma boa saúde financeira do empreendimento,
# é fundamental que esses pagamentos sejam feitos de forma regular e pontual. No entanto, sabemos que nem sempre isso acontece. Pensando nisso, proponho um
# desafio de tratamento de dados com o objetivo de analisar o atraso no pagamento de aluguéis no condomínio fictício de alguns moradores.
```

```python
dados_condominio['datas_combinadas_pagamento'] = pd.to_datetime(dados_condominio['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_condominio['datas_de_pagamento'] = pd.to_datetime(dados_condominio['datas_de_pagamento'], format = '%d/%m/%Y')
dados_condominio.info()
```

```python
dados_condominio['Atrasos'] = (dados_condominio['datas_de_pagamento'] - dados_condominio['datas_combinadas_pagamento']).dt.days
```

```python
dados_condominio
```

```python
media_atraso = dados_condominio.groupby(['apartamento'])['Atrasos'].mean()
media_atraso
```
